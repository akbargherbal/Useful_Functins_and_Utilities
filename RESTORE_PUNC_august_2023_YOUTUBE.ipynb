```python
dir_path = '/content/gdrive/MyDrive/Colab_Notebooks/Spectrum_Management/CC_FILES/'
output_folder = '/content/gdrive/MyDrive/Colab_Notebooks/Spectrum_Management/PUNCT_FILES'
```

```python
import os
```

```python
files_path = []
for root, folders, files in os.walk(dir_path):
  for file in files:
    if file.endswith('.pkl'):
      path_file = os.path.join(root, file)
      files_path.append(path_file)

```

```python
files_path
```

```python
from rpunct import RestorePuncts
import pandas as pd
```

```python
rpunct = RestorePuncts()
```

```python
def restore_punc(text):
    try:
        result = rpunct.punctuate(text)
        return result
    except Exception as e:
        print(e)
        return f'ERROR_PUNCT: {text}'
```

```python
for (idx, file) in enumerate(files_path):
    print(f'Processing file {idx+1} of {len(files_path)}...')
    file_name = file.split('/')[-1]


    df = pd.read_pickle(file)
    print(f'File {file} is {len(df)} mintues long')

    print(f'{output_folder}/PUNCT_{file_name}')

    df['PUNCT'] = df['text'].apply(lambda x: restore_punc(x))



    df.to_pickle(f'{output_folder}/PUNCT_{file_name}', protocol=4)
    print(f'Finised file {idx+1} of {len(files_path)}')
    print('--------')
```
