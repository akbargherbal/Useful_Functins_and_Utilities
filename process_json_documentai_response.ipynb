```python
with open('./document_ai_sample-0.json', encoding='utf-8') as jf:
    data = json.load(jf)
```

```python
import pandas as pd
```

```python
all_text = data['text']
```

```python
list_para_idx = []
for page in data['pages']:
    for (k,v) in page.items():
        if k == 'paragraphs':
            for (i) in v:
                for k1,v1 in i.items():
                    if k1 == 'layout':
                        print(k1, type(v1))
                        print(i[k1].keys())
                        print(type(i[k1]['textAnchor']['textSegments']))
                        for ii in i[k1]['textAnchor']['textSegments']:
                            print(f"LEVEL textSegments {type(ii)}")
                            for k2, v2 in ii.items():
                                print(k2, type(v2))
                                start_idx = int(ii.get('startIndex', "0").strip())
                                print(F'START_INDEX: {start_idx}')
                                end_idx = int(ii.get('endIndex'))
                                print(F'END_INDEX: {end_idx}')
                            list_para_idx.append((start_idx, end_idx))
                        print('-'*50)
                        
```

```python
all_paragraphs = ''.join(
[
all_text[i[0]:i[1]]

for i in list_para_idx

]
)
```

```python
print(all_paragraphs)
```

```python
with open('para_text.txt', encoding='utf-8', mode='w') as f:
    f.write(all_paragraphs)
```
