```python
list_para_idx = []
for page in data['pages']:
    for (k,v) in page.items():
        if k == 'blocks':
            for (i) in v:
                for k1,v1 in i.items():
                    if k1 == 'layout':
                        print(k1, type(v1))
                        print(i[k1].keys())
                        print(type(i[k1]['textAnchor']['textSegments']))
                        for ii in i[k1]['textAnchor']['textSegments']:
                            print(f"LEVEL textSegments {type(ii)}")
                            for k2, v2 in ii.items():
                                print(k2, type(v2))
                                start_idx = int(ii.get('startIndex', "0").strip())
                                print(F'START_INDEX: {start_idx}')
                                end_idx = int(ii.get('endIndex'))
                                print(F'END_INDEX: {end_idx}')
                            list_para_idx.append((start_idx, end_idx))
                        print('-'*50)
                        
```

```python
df = pd.DataFrame(data={'IDX': list_para_idx})
```

```python
df['TEXT'] = df.IDX.apply(lambda x: all_text[x[0]: x[1]]  ) 
```

```python
df['LEN'] = df['TEXT'].apply(lambda x: len(x.split()))
```

```python
df['TEXT_NEW'] = df.apply(lambda x:  re.sub('\n+', ' ', x['TEXT']) if (x['LEN'] >12) else x['TEXT'], axis=1)
```

```python
df['TEXT'.split()].to_csv('NEW_TEXT.txt', encoding='utf-8', index=False, header=False)
```

```python
list_text = [re.sub(' +', ' ', i).strip() for i in list(df.TEXT_NEW)]
```
